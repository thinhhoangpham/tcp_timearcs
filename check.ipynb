```python
import pandas as pd

input_path = 'set10_full_min_matched_attacks.csv'

df = pd.read_csv(input_path)
df.head()
```

```python
# Just convert microsecond timestamps to minutes
df['timestamp'] = (df['timestamp'] // 1_000_000 // 60).astype(int)
df.head()
```

```python
# collapse duplicates on all columns except 'count', summing 'count'
before = len(df)
group_cols = [c for c in df.columns if c != 'count']
df = df.groupby(group_cols, as_index=False)['count'].sum()
after = len(df)
print(f'Collapsed {before - after} duplicate groups; now {after} rows.')

df.head()
```

```python
# write dataframe to a compressed CSV file
out_path = input_path.replace('.csv', '_out.csv')
df.to_csv(out_path, index=False)
print(f'Wrote {len(df)} rows to {out_path}')
```
